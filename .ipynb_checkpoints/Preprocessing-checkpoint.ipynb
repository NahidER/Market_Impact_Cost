```python
# Convert to date & time
trade['Date'] = pd.to_datetime(trade['Date'])
trade['Time'] = pd.to_datetime(trade['Time'])
lob['Date'] = pd.to_datetime(lob['Date'])
lob['Time'] = pd.to_datetime(lob['Time'])
```

```python
# Copy of Trade
trade = trade.copy()
lob = lob.copy()
```

```python
# Grouping trades with equal time & date (sum of their volume)
trade1 = trade.groupby(['Date','Time'], as_index=False).agg({'Id':'last','Volume':'sum'})
```

```python
# Grouping trades with equal time & date (vwap of prices)
trade2 = trade.groupby(['Date','Time']).apply(lambda x: np.average(x.Price, weights=x.Volume))

# DataFrame 
vwap = trade2.values
vwap = pd.DataFrame(vwap, columns=['price'])
vwap
```

```python
# Final trades 
trade1 ['Price'] = vwap
```

```python
trade1
```

```python
# Merge Trade & Lob based on Time & Date
trlo = pd.merge(trade1, lob, how='left', on=['Date','Time'])
trlo = trlo [(trlo.A1_p2.notnull()) & (trlo.B1_p.notnull())]
trlo.head()
```

```python
# Lob Shifting
lob_shifted = lob.copy()
for col in lob.columns:
    if col == 'Date' or col == 'Time':
        continue
    lob_shifted[col] = lob_shifted[col].shift(1)
```

```python
# Merge shifted lob with trade
trlo = pd.merge(trlo, lob_shifted, how='left', on=['Date','Time'], suffixes=('','_sf'))
trlo.head()
```

```python
# calculating Mid-Point
trlo['mid-point'] = (trlo['B1_p_sf'] + trlo['A1_p2_sf']) / 2
```

```python
trlo.columns
trlo.head()
```

```python
# Trade side detection
ls = [] 
for i in range(len(trlo)):
    if (trlo.iloc[i]['Price'] >= trlo.iloc[i]['mid-point']):
        ls.append(1)
    elif (trlo.iloc[i]['Price'] < trlo.iloc[i]['mid-point']):
        ls.append(-1)
    else :
        print("not in code")
trlo['trade_side'] = ls
```

```python
# Calculating Istar
trlo['I_star'] = np.log(trlo['Price'] / trlo['mid-point'])
```

```python
trlo.head()
```

```python
trlo.isnull().sum()
```

```python
# Deviding trades into 15min intervals
main=  []
for date in trlo.Date.unique():
    low = pd.to_datetime('09:00:00')
    for j in range(14):
        d = trlo[(trlo['Date'] == date) & (trlo['Time'] >= low) & (trlo['Time'] < low + pd.Timedelta('15 min'))]
        #print(d)
        low = low + pd.Timedelta('15 min')
        main.append(d)
```

```python
# Average Daily Volume
vol = []
for date in trlo.Date.unique():
    tr = trlo[(trlo['Date'] == date)]
    voll = sum(tr['Volume'])
    vol.append(voll)
vol = pd.DataFrame(vol, columns=['Vol'])
```

```python
# Calculating Volume Imbalance for each interval
Imb = []
for i in range(len(main)):
    buy_vol = main[i][(main[i]['trade_side'] == 1)]
    sell_vol = main[i][(main[i]['trade_side'] == -1)]
    imbalance = (sum(buy_vol['Volume']-sum(sell_vol['Volume'])))
    Imb.append(imbalance)
Imb = pd.DataFrame(Imb, columns=['Volume Imbalance'])
```

```python
Imb.head()
```

```python
# Calculating Average of I_star cost
Cost = []
for i in range(len(main)):
    cost = np.mean(main[i]['I_star'])
    Cost.append(cost)
    
Cost = pd.DataFrame(Cost, columns=['Istar cost'])
Cost.head()
```

```python
# Calculating Q/ADV
ADV = int(np.mean(vol))
Q_adv = (np.abs(Imb['Volume Imbalance']))/(ADV)
Q_adv
```
